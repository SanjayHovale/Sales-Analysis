QUE4:What 2 products are sold together most?


```python
dup=df[df['Order ID'].duplicated(keep=False)]
dup.head()
```

```python
dup.shape
```

```python
dup['grouped']=dup.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
dup.head()
```

```python
dup=dup[['Order ID','grouped']].drop_duplicates()
dup.head()
```

```python
dup.shape
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in dup['grouped']:
  row_list=row.split(',')
  count.update(Counter(combinations(row_list,2)))
print(count)  
```

```python
for key,v in count.most_common():
  print(key,v)
```

ANS4:('iPhone', 'Lightning Charging Cable') 1005 are sold together most
